```python
#Part 1 — value_counts()
import pandas as pd
employees = {
    "Name": ["Asha", "Rahul", "Kiran", "Meera", "Arjun", "Sneha"],
    "Department": ["IT", "HR", "IT", "Sales", "Sales", "IT"],
    "Salary": [50000, 40000, 55000, 45000, 48000, 60000]
}
df = pd.DataFrame(employees)

print(df)
#How many employees are there in each department
print(df.groupby('Department')['Name'].count())
```

```python
#instead Pandas gives us a faster tool for frequency questions:
print(df['Department'].value_counts()) #it checks department col and counts the number of distinct vaules
```

```python
#Part 2 — unique() vs nunique()
#Question like “Which departments exist in our company?”
#use unique()
print(df['Department'].unique()) #Gives actual unique vaules

#Question like “How many different departments exist?”
#use nunique()
print(df['Department'].nunique()) #Gives the number of unique values
```

```python
#Pandas Final Case Study — Employee Performance Analysis
employees = {
    "EmployeeID": [101, 102, 103, 104, 105, 106],
    "Name": ["Asha", "Rahul", "Kiran", "Meera", "Arjun", "Sneha"],
    "Department": ["IT", "HR", "IT", "Sales", "Sales", "IT"]
}

performance = {
    "EmployeeID": [101, 102, 103, 104, 105, 106],
    "Projects": [8, 5, 9, 7, 6, 10],
    "Rating": [4.2, 3.8, 4.5, 4.0, 3.7, 4.8],
    "Salary": [50000, 40000, 55000, 45000, 48000, 60000]
}
employees_df = pd.DataFrame(employees)
performance_df = pd.DataFrame(performance)

# Task 1(Merge both datasets using EmployeeID.)

merged_df = pd.merge(employees_df,performance_df,on='EmployeeID')

# Task 2(Print the first 3 rows of the merged dataset.)

print(merged_df.head(3))

# Task 3(Find how many employees belong to each department using value_counts().)

print(merged_df['Department'].value_counts())

# Task 4(Print all unique departments.)

print(merged_df['Department'].unique())

# Task 5(Find the number of unique departments.)

print(merged_df['Department'].nunique())

# Task 6(Find the average salary for each department using groupby().)

avg_sal = merged_df.groupby('Department')['Salary'].mean()
print(avg_sal)

# Task 7(Find employees whose rating is greater than 4.0.)

print(merged_df[merged_df['Rating']>4.0])

# Task 8(Sort employees by salary from highest to lowest.)

print(merged_df.sort_values('Salary', ascending=False))
# Task 9(Print only the highest-paid employee using .head(1).)

print(merged_df.sort_values('Salary', ascending=False).head(1))

#Find the department with the highest average rating.
merged_df.groupby("Department")["Rating"].mean()
#Based on the Average Rating's of each Departments, IT Department has a better overall rating, followed by Sales and HR.
```
