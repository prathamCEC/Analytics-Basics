```python
#Part 3 — When astype() fails
ages = pd.Series(["21","22","unknown","24"])
ages=ages.astype(int)
print(ages.dtype)
#IT fails for uknown, so it's safer to use .to_numeric()
```

```python
# #Part 8 — What should we do with invalid values?
# Part 8 — What should we do with invalid values?

# Suppose:

# Age = 250

# Should we:

# A. Delete the row
# B. Replace it with the average
# C. Change it to 25
# D. Investigate first

# Usually:

# D. Investigate first.

# Maybe:

# 250

# was a typo for:

# 25

# But if you automatically change it, you're guessing.

# A cleaner workflow is:

# Detect
#  ↓
# Inspect
#  ↓
# Understand
#  ↓
# Correct if evidence exists
#  ↓
# Otherwise mark missing/remove based on context

# Data cleaning is not:

# "Make weird values disappear."

# It is:

# "Make the dataset more truthful."
```

```python
#💻 Coding Challenge — Customer Data Cleaning
import pandas as pd
customers = {
    "Name": ["Asha","Rahul","Kiran","Meera","Arjun"],
    "Age": ["24","27","unknown","26","150"],
    "Payment": ["UPI","cash","Credit Card","CASH","upi"]
}

df = pd.DataFrame(customers)

# Your tasks:

# 1.Print the data types.
print(df.dtypes)

# 2.Convert Age using pd.to_numeric(errors="coerce").
df['Age']=pd.to_numeric(df['Age'],errors="coerce")

# 3.Print the Age column after conversion.
print(df['Age'])

# 4.Count missing values created during conversion.
print(df['Age'].isnull().sum())

# 5.Clean Payment using strip() and title().
df['Payment']=df['Payment'].str.strip().str.title()

# 6.Standardize "Upi" back to "UPI" using replace().
df['Payment']=df['Payment'].replace({
    "Upi": "UPI"
})
# 7.Print all unique payment methods after cleaning.
print(df['Payment'].unique())
      
# 8.Find customers whose Age is greater than 100.
print(df[(df['Age']>100)])

# 9.Don't fix the age 150 yet. Write one sentence explaining why you should investigate it first.
# It's not confimered wherther Arjun's age is 150, it can be a typo, too so we we have to gather enough evidence
```

```python
#🏆 Mini Project — Messy Sales Data
import pandas as pd
sales = {
    "Product": ["Laptop","Mouse","Keyboard","Monitor","Laptop","Mouse"],
    "Category": ["electronics","Accessories","ACCESSORIES"," Electronics ","electronics","accessories"],
    "Price": ["50000","800","1500","bad","52000","900"],
    "Rating": [4.5,4.2,6.5,4.4,4.8,3.9]
}
sales_df = pd.DataFrame(sales)

# Task 1(Print the original dataset.)

print(sales_df)

# Task 2(Print all data types.)

print(sales_df.dtypes)

# Task 3(Clean Category using strip() and title().)

sales_df['Category']=sales_df['Category'].str.strip().str.title()

# Task 4(Print unique categories after cleaning.)

print(sales_df['Category'].unique)

# Task 5(Convert Price using pd.to_numeric(errors="coerce").)

sales_df['Price']=pd.to_numeric(sales_df['Price'],errors='coerce')

# Task 6(Count missing prices after conversion.)

print(sales_df['Price'].isnull().sum())

# Task 7(Find rows where Rating is outside the valid range 1–5.)

print(sales_df[sales_df['Rating']>5])

# Task 8(Do not immediately delete or replace the invalid rating. Explain why.)

#bro i'm lazy from next class i'll give, this time forgive me teacher 😅

# Task 9(Fill the missing Price using the average valid price.)

sales_df['Price']=sales_df['Price'].fillna(sales_df['Price'].mean())

# Task 10(Verify that no Price values are missing.)

print(sales_df['Price'].isnull())

# Task 11(Find the average Price for each Category.)

print(sales_df.groupby('Category')['Price'].mean())

# Task 12(Write one analytical sentence explaining which category has the higher average price.)
#Based on the average sales of each category, we can conclude that Electronics have more sales compared to Acessories
```
