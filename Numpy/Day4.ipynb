```python
#Final Mini Project 
#Student Analytics System

import numpy as np

marks = np.array([
    [78,85,90],
    [65,70,68],
    [92,88,95],
    [55,60,58],
    [80,75,85],
    [92,88,95]
])#col->math,science,english & row->students

#Task1(Find average marks per subject.)

print(np.mean(marks,axis=0))

#Task2(Find topper total marks.)

print(np.max(np.sum(marks,axis=1)))

#Task3(Find student indexes whose average > 80.)

print(np.where(np.mean(marks,axis=1)>80))

#Task4(Find highest score in each subject.)

print(np.max(marks,axis=0))

#Task5(Sort in descending)

totals = np.sum(marks,axis=1)
print(totals)#default order
print(np.sort(totals)[::-1])#desecending order

#Task6(Find all unique total scores.)

print(np.unique(np.sum(marks,axis=1)))
```
