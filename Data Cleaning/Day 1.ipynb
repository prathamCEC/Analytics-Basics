```python
#Part 6 — Cleaning inconsistent text
# Look at our cities:
# Bangalore
# mangalore
# Mysore
# Bangalore 
# MANGALORE
print(df['City'].unique()) #we notice all the vaules as unique(case are different)
#steps of Cleaning inconsistet text

#Step 1: Remove extra spaces with .str.strip()
df['City'] = df['City'].str.strip()
#Think of strip() as a scissors it will removes spaces from the beginning and at the end
print(df['City'])
#Note it does'nt removes space inside meaningful text

#Step 2: Standardize capitalization
df['City']=df['City'].str.title()
print(df['City'])
#After applying  .str.title()
#mangalore  → Mangalore
# MANGALORE  → Mangalore
# bangalore  → Bangalore

#inShort cleaning is the combination of both
#df['City'] = df['City'].str.strip().str.title()

# Read it from left to right:
# Take the City column → remove surrounding spaces → convert text to title case.
# This is called method chaining.
```

```python
# #Part 7 — Missing values: don't blindly fill them
# Our data has:

# Bob → Age missing
# David → Salary missing

# Could we use:

# df = df.dropna()

# Yes.

# But that removes both Bob and David completely.

# We lose:

# their names
# cities
# all other valid information

# Maybe that's too expensive.

# Instead, we can fill numerical missing values with the average:

df["Age"] = df["Age"].fillna(df["Age"].mean())

# Read it in English: Find the average age of the available people and use it for the missing age.
# Same idea:
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
print(df)
# But why use the mean?

# Suppose salaries are:
# 40000
# 50000
# 60000
# Missing
# Average: 50000

# Using 50000 is often more reasonable than:
# 0
# because a missing salary doesn't mean the person earns nothing.

# Important: The mean is not always the best choice.

# Later we'll discuss:
# Mean
# Median
# Mode
# Dropping
# Business-specific values

# For now, the main lesson is: Missing doesn't mean zero.
```

```python
#🏆 Mini Project — Messy Employee Data Cleanup
employees = {
    "Name": ["Asha","Rahul","Kiran","Meera","Asha","Arjun"],
    "Department": ["IT"," hr ","SALES","it","IT","Sales"],
    "Age": [24,27,None,26,24,30],
    "Salary": [50000,45000,55000,None,50000,60000]
}
emp_df = pd.DataFrame(employees)

# Task 1(Print the original dataset.)

print(emp_df)

# Task 2(Print its shape.)

print(emp_df.shape)

# Task 3(Count missing values.)

print(emp_df.isnull().sum())

# Task 4(Count duplicate rows.)

print(emp_df.duplicated().sum())

# Task 5(Remove duplicate rows.)

emp_df=emp_df.drop_duplicates()
print(emp_df)

# Task 6(Clean the Department column using strip() and title().)

emp_df['Department']=emp_df['Department'].str.strip().str.title()
print(emp_df['Department'])

# Task 7(Fill the missing Age with the average age.)

emp_df['Age'] = emp_df['Age'].fillna(emp_df['Age'].mean())
print(emp_df['Age'])

# Task 8(Fill the missing Salary with the average salary.)

emp_df['Salary'] = emp_df['Salary'].fillna(emp_df['Salary'].mean())
print(emp_df['Salary'])

# Task 9(Verify that no missing values remain.)

print(emp_df.isnull())

# Task 10(Print the final cleaned dataset.)

print(emp_df)
```

```python
#⭐ Analyst Task
# Which department has the highest average salary?
print(emp_df.groupby('Department')['Salary'].mean())
#Based on the average salary's of each department, Sales depratment have the heighest average salary, followed by IT It and Hr
```
