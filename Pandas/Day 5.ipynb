```python
#Our dataset
import pandas as pd

students = {
    "Name": ["John", "Alice", "Bob", "David", "Eva", "Sam"],
    "Section": ["A", "A", "B", "B", "A", "B"],
    "Math": [78, 92, 65, 80, 95, 75],
    "Science": [85, 88, 70, 90, 91, 82]
}

df = pd.DataFrame(students)
print(df)
```

```python
#Understanding groupby()
#imange 2 backets A section, and B section
df.groupby("Section")#it doens't calculate anything, instead it groups the students based on section
```

```python
#Average Marks by Section
#Find avg Math marks in each sec
print(df.groupby('Section')['Math'].mean())
```

```python
#Challenge 1
print(df.groupby('Section')['Science'].max())
```

```python
#Count Students
#"How many students are in each section?"
print(df.groupby('Section')['Math'].count())
```

```python
#Challenge 3
print(df.groupby('Section')['Math'].sum())
```

```python
#Multiple Aggregations
#Give me average, highest, and lowest Math marks for each section.
print(
    df.groupby('Section')['Math'].agg(
        ["mean","max","min"]
    )
)
#.agg()-> contains the aggrigate function, so just by creating list of aggregate funcyions["mean","max","min"] we can use thoes in 'Math'
```

```python
#Challenge 3
print(df.groupby("Section")["Science"].min())
```
