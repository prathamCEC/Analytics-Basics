```python
#Mini Project: Student Dataset Cleanup
data = {
    "Name":["John","Alice","Bob","David"],
    "Math":[78,None,65,80],
    "Science":[85,88,None,90],
    "English":[90,95,68,None]
}

df = pd.DataFrame(data)
print(df)
# Task 1(Print the first 2 rows.)

print(df.head(2))

# Task 2(Print dataset information.)

print(df.info())

# Task 3(Find missing values in each column.)

print(df.isnull())
print(df.isnull().sum())

# Task 4(Remove rows with missing values.)

df=df.dropna()

# Task 5(Fill missing Math marks with:0)

df['Math'] = df['Math'].fillna(0)

# Task 6(Print the cleaned dataset.)

print(df)
```
