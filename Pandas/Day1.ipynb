```python
#Step 2:Creating a Series
#Series ia a single col
marks=pd.Series([78,92,65,80])
print(marks)
```

```python
#Step3: Creating DataFrame

data={
    "name":["Jhon","Alice","Bob"],
    "Math":[78,92,65],
    "Science":[85,88,70]
}

df=pd.DataFrame(data)
print(df)
```

```python
#in Numpy to exract a single col  we write
#marks[:,0]
#in Pandas just write
#df["Math"]

# df[:2, :2]
```

```python
#Selecting Columns
print(df['Math'])
```

```python
#average marks
print(df['Math'].mean())#single #col inside 1 square brackets  
```

```python
#Selecting Multiple Columns

Multiple_df=df[['name','Math','Science']] #col inside 2 square brackets  
print(Multiple_df)
```

```python
#Mini Project
students={
    "Name":["John","Alice","Bob","David"],
    "Math":[78,92,65,80],
    "Science":[85,88,70,90],
    "English":[90,95,68,85]
}

df=pd.DataFrame(students)

#Task 1(Print the entire DataFrame.)

print(df)

#Task2(Print only the Math column.)

print(df['Math'])

#Task3(Find average Math marks.)

print(df['Math'].mean())

#Task4(Find highest Science marks.)

print(df['Science'].max())

#Task5(Print the shape.)

print(df.shape)

#Task 6(Print all column names.)

print(df['Name'])

#Task 7(use:)

print(df.describe())
#English has the highest mean
```
