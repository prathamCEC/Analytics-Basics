```python
#Creating Arrays
import numpy as np
marks=np.array([10,20,30,40,50])
print(marks) #no commans
```

```python
#python List vs Array

marks=[10,20,30,40,50]
new_marks=[]
for i in marks:
    new_marks.append(i+5)
print(new_marks)
arr=np.array([10,20,30,40,50])
print(arr+5)
```

```python
#indexing same as python list
arr=np.array([10,20,30,40,50])
print(arr[0])
print(arr[-1])
```

```python
#slicing same as python list
arr=np.array([10,20,30,40,50])
print(arr[0:3])
```

```python
#Vectorized operations
arr=np.array([10,20,30,40,50])
print(arr+1)
print(arr-2)
print(arr*3)
print("Constantly used in analytics")
```

```python
#Statistics
arr=np.array([10,20,30,40,50])
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(np.sum(arr))
```

```python
#Mini Challenge
import numpy as np
scores = np.array([70, 80, 90, 60, 100])
#1. Add 10 marks to everyone
print(scores+10)

#2.Find Average
print(np.mean(scores))

#3.Find highest score
print(np.max(scores))

#4.Find lowest score
print(np.min(scores))
```

```python
#Reshaping
arr=np.array([1,2,3,45,6])
print(arr.shape)
#reshape is used when you want to represent a array in 2-D
arr=np.array([1,2,3,4,5,6])
new_arr=arr.reshape(3,2)#how many (rows, col)
print(new_arr)#rule to remind while reshaping array row*col==number of array elements
```
