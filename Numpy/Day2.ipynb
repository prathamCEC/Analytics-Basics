```python
import numpy as np
marks = np.array([
    [70,80,90],
    [60,85,95],
    [90,75,88]
])
print(marks)
#before using axis
sci_marks=marks[:,1]
print(np.sum(sci_marks))
```

```python
#axis=0(column wise)Sum
print(np.sum(marks,axis=0)) #70+60+80
```

```python
#axis=1(row wise)sum
print(np.sum(marks,axis=1))#70+80+90
```

```python
#mean with aixs
print(np.mean(marks,axis=0))#col wise
print(np.mean(marks,axis=1))#row wise
```

```python
#maximum with axis
print(np.max(marks,axis=0))
print(np.max(marks,axis=1))
```

```python
#Analytics example
sales=np.array([
    [100,120,130],#month->col,  store->row
    [150,170,180],
    [200,220,250]
])
#sales per month
print(np.sum(sales,axis=0))
#sales per store
print(np.sum(sales,axis=1))
```

```python
#coding Task
sales=np.array([
    [100,200,300],
    [150,250,350],
    [200,300,400]
])
#1.print shape
print(sales.shape)

#2.Print Tanspose
print(sales.T)

#3.Total sales per month
print(np.sum(sales,axis=0))

#4.Total sales per sotre
print(np.sum(sales,axis=1))

#5.Average sales per month

print(np.mean(sales,axis=0))

#6.highest sles in each month

print(np.max(sales,axis=0))
```

```python
#Coding challenge
import numpy as np
scores = np.array([55, 80, 45, 90, 70, 30])

#1.all scores greater than 60

print(scores[scores>60])

#2. Print all scores less than 50.

print(scores[scores<50])

#3.Find the average of scores greater than 60.

print(np.mean(scores[scores>60]))

#4.Find the highest score among scores greater than 60

print(np.max(scores[scores>60]))
```
