```python
#Boardcasting
import numpy as np
marks = np.array([70, 80, 90])
#add five to each element
print(marks+5)#boardcasting [70+5 85+5 90+5]
```

```python
#Analytics example
#Suppose GST increased by 18%.
sales=np.array([1000, 2000, 3000])
print(sales*1.8)#sales with 18% gst
```

```python
#Boardcasting between array
A=np.array([10,20,30])
B=np.array([1,2,3])
print(A+B) #A+B=[10+1 20+2 30+3]
```

```python
#When Broadcasting Fails
A=np.array([10,20,30])
B=np.array([1,2])
print(A+B) #while broadcasting we have to ensure that array contains same number of rows and col
```

```python
#Advanced Filtering
#AND condition
#Print all scores greater than 50. and less than 90 
scores=np.array([40,55,70,85,95])
#same as normal filtering but contians logical operations
print(scores[(scores>50)&(scores<90)])#conditions inside ()
```

```python
#OR condition
#Print all scores less than 50. or greater than 90 
print(scores[(scores<50)|(scores>90)])
```

```python
#analytics example
marks=np.array([45,67,89,32,95,78])
print(marks[marks>=75])
```

```python
#challenge 2
scores = np.array([20,40,60,80,100])

print(scores[(scores >= 40) & (scores <= 80)])
```

```python
#Normal Multiplication
#Element-wise multiplicaton
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print(A * B)##[[1*5 2*6]
                ##[3*7 4*8]]
```

```python
#Matrix Multiplication
print(A@B)
```

```python
#Mini Project
#Student Performance Analyzer
import numpy as np

marks = np.array([
    [78,85,90],
    [65,70,68],
    [92,88,95],
    [55,60,58],
    [80,75,85]
])#col->math,science,english & row->students

#Task1(Find average marks for each subject.)

print(np.mean(marks,axis=0))

#Task2(Find total marks for each student.)

print(np.sum(marks,axis=1))

#Task3(Find the topper's total marks.)

print(np.max(np.sum(marks,axis=1)))

#Task4(Find students whose average marks are above 80.)

student_averages=np.mean(marks, axis=1)
print(student_averages[student_averages>80])
```
