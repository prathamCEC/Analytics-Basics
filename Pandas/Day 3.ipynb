```python
import pandas as pd

students ={
    "Name":['Jhon','Alice','Bob','David','Eva'],
    "Math":[78,92,65,80,95],
    "Science":[85,88,70,90,91],
    "English":[90,95,68,85,97]
}
df = pd.DataFrame(students)
print(df)
```

```python
#Adding a New column
#Adding Total_Marks col
#total marks of each students eg(Jhon->78+85+90 and so on)
#instead of that we can create a new column.
df['Total_Marks'] = df['Math']+df['Science']+df['English']
print(df)
```

```python
#Why create new Col
#Example for finding the topper of the class we can use max() to Total_Marks itself
print(f"Topper:{df['Total_Marks'].max()}")
```

```python
#Adding Average col
df['Average'] = df['Total_Marks']/3
print(df)
```

```python
#Updating a value
df.loc[2,'Math'] = 75 #or df.iloc[2,1] = 75
print(df)
#loc mean labels
```

```python
#Challenge 2
df.loc[0,"Science"]=100
print(df.loc[0])
```

```python
#Sorting Data
#Suppose you want students ranked by total marks
#Ascending order
print(df.sort_values("Total_Marks"))
print()
#Descending Order
print(df.sort_values("Total_Marks",ascending=False))
```
