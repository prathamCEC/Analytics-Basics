```python
#Understanding iloc #Integer Location
import pandas as pd
students = {
    "Name": ["John", "Alice", "Bob", "David"],
    "Math": [78, 92, 65, 80],
    "Science": [85, 88, 70, 90],
    "English": [90, 95, 68, 85]
}
df=pd.DataFrame(students)
print(df)
#the numbers at the left most is called as indexes
#Python List Index
names=["Jhon","bob","Mat"]
print(names[0],names[-1])
```

```python
#instead of print(names[0],names[-1]) we can use ->
print(df.iloc[0]) #0th row details 0   John    78       85       90
```

```python
#Visualizing iloc
print(df.iloc[1,2]) #->1st row, 2 col-> vaule which is 88
```

```python
#Filtering 
#students scored more than 80 in Math
print(df[df['Math']>80])
print()
print(df["Math"]>80)#Boolean Masking
```

```python
#Real Analyst Example
emp={
    "E_name":["A","B","C"],
    "sal":[30000,60000,45000]
}
df=pd.DataFrame(emp)
print(df[df['sal']>50000])
```

```python
#Multiple Conditions in Filtering
#print Students having >70 marks in Math AND >85 marks in Science
print(df[(df['Math']>70)&(df['Science']>85)])
```

```python
import pandas as pd

students = {
    "Name": ["John", "Alice", "Bob", "David", "Eva"],
    "Math": [78, 92, 65, 80, 95],
    "Science": [85, 88, 70, 90, 91],
    "English": [90, 95, 68, 85, 97]
}

df = pd.DataFrame(students)

print(df)
```

```python
#Challenge 1
print(df.iloc[2])
```

```python
#challenge 2
print(df.iloc[1,3])
```

```python
#Challenge 3
print(df[df['Math']>80])
```

```python
#Challenge 4
print(df[(df['Science']>85)&(df['English']>90)])
```

```python
#coding Exercise
#Task 1(print the first row using iloc)

print(df.iloc[0])

#Task 2(Print rows 1 to 3.)

print(df.iloc[1:4])

#Task 3(Print only students whose Math marks are greater than 75.)

print(df[df['Math']>75])

#Task 4(Print students whose English marks are less than 90.)

print(df[df['English']<90])

#Task 5(Math > 70 AND Science > 85.)

print(df[(df['Math']>70 )&(df['Science']>85)])
```

```python
#Mini Project
#Princi need report
print(df)
#Task1(Find students who scored more than 90 in Math.)

print(df[df['Math']>90])

#Task2(Find students who scored more than 85 in all three subjects.)

print(df[(df['Math']>85)&(df['Science']>85)&(df['English']>85)])

#Task3(Find students who scored below 70 in any subject.)

print(df[(df['Math']<70)|(df['Science']<70)|(df['English']<70)])

#Task4(Print the details of the student at index 3 using iloc.)

print(df.iloc[3])

#Task5(Print only the Name and Math columns of students whose Math marks are above 80.)

print(df[df['Math']>80][['Name','Math']])
```
